```python
import duckdb
conn = duckdb.connect("/content/drive/MyDrive/BigData/1brc/measurements.db") # 메모리 대신 디스크에 DB를 저장
conn.execute(f'''CREATE TABLE if not exists temp AS
                SELECT * FROM read_csv_auto('/content/drive/MyDrive/BigData/1brc/measurements.txt', delim=';', header=False, columns = {{'location':'VARCHAR','temperature':'float'}})''')
conn.close()
```
